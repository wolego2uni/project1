# **INTRODUCTION TO TENSORFLOW** 

#### **WEEK 1 TENSORFLOW DEVELOPERS CERTIFICATE TRAINING** 

#### **IMPORTS** 

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## **IMPORTING DATA** 

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

#A simple numpy array of x and y
```

## **MODEL** 

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
#MEAN SQUARED ERROR' for the loss and 'STOCHASTIC GRADIENT DESCENT' for the optimizer
```

## **TRAINING THE SIMPLE NEURAL NETWORK** 

```python
model.fit(xs, ys, epochs=600)
```

## **TESTING** 

```python
print(model.predict([10.0]))
```
